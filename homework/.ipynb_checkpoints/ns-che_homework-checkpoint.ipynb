```python
dataset = pd.read_csv('./data/accepted_2007_to_2018Q4.csv')
```

```python
dataset.head()
```

## loan_amnt: 대출 총액
## term 대출 기간
## loan_status 대출 상태
## grade 대출 등급
## int_rate 대출 이자율


```python
data = dataset[['loan_amnt','loan_status', 'grade', 'int_rate', 'term']]
```

```python
loan_value36 = data.loc[data['term']==' 36 months', 'loan_amnt'].sum()
loan_value60 = data.loc[data['term']==' 60 months', 'loan_amnt'].sum()

print(f"36개월 대출 대출총액: {int(loan_value36)}")
print(f"60개월 대출 대출총액: {int(loan_value60)}")
```

```python
data['loan_status'].unique()

bad_status = ['Charged Off', 'Default', 'Does not meet the credit policy. Status:Charged Off','In Grace Period','Late (16-30 days)','Late (31-120 days)']

print("정상 대출자 등급 파악")
print(data.loc[~(data['loan_status'].isin(bad_status)), 'grade'].value_counts().sort_index())
print()
print("불량 대출자 등급 파악")
print(data.loc[data['loan_status'].isin(bad_status), 'grade'].value_counts().sort_index())

```

```python
print(f"대출총액과 대출 이자율 간의 상관관계: {data['loan_amnt'].corr(data['int_rate']):.3f}")
```
